3. Escribir una función que reciba un número entero positivo y devuelva su factorial.

```python
def factorial(n):
  if n < 0:
    raise ValueError("El factorial no está definido para números negativos")
  elif n == 0:
    return 1
  else:
    resultado = 1
    for i in range(1, n + 1):
      resultado *= i
    return resultado

# Ejemplo de uso
print(factorial(4)) # Salida: 24
```

5. Escribir una función que calcule el área de un círculo y otra que calcule el volumen de un cilindro usando la primera función.

```python
import math

def calcular_area_circulo(radio):

  return math.pi * (radio ** 2)

def calcular_volumen_cilindro(radio, altura):

  area_base = calcular_area_circulo(radio)
  return area_base * altura

# Ejemplo de uso
radio_ejemplo = 5
altura_ejemplo = 10

area = calcular_area_circulo(radio_ejemplo)
print(f"El área del círculo con radio {radio_ejemplo} es: {area:.2f}")

volumen = calcular_volumen_cilindro(radio_ejemplo, altura_ejemplo)
print(f"El volumen del cilindro con radio {radio_ejemplo} y altura {altura_ejemplo} es: {volumen:.2f}")
```

6. Escribir una función que reciba una muestra de números en una lista y devuelva su media.

```python
def calcular_media(muestra):
  if not muestra: # Manejar el caso de una lista vacía
    return 0
  return sum(muestra) / len(muestra)

# Ejemplo de uso
numeros = [1, 2, 3, 4, 5]
media = calcular_media(numeros)
print(f"La media de la lista es: {media}")
```

7. Escribir una función que reciba una muestra de números en una lista y devuelva otra lista con sus cuadrados.

```python
def calcular_cuadrados(lista_numeros):
  cuadrados = [] # Se inicializa una lista vacía para almacenar los resultados
  for numero in lista_numeros: # Se itera sobre cada número en la lista de entrada
    cuadrados.append(numero ** 2) # Se calcula el cuadrado y se añade a la nueva lista
  return cuadrados # Se devuelve la lista con los cuadrados
```

8. Escribir una función que reciba una muestra de números en una lista y devuelva un diccionario con su media, varianza y desviación típica.

```python
import math

def calcular_estadisticas(muestra):
    n = len(muestra)
    if n == 0:
        return {}

    # Calcular la media
    media = sum(muestra) / n

    # Calcular la varianza
    varianza = sum([(x - media) ** 2 for x in muestra]) / n

    # Calcular la desviación típica
    desviacion_tipica = math.sqrt(varianza)

    return {
        'media': media,
        'varianza': varianza,
        'desviacion_tipica': desviacion_tipica
    }

# Ejemplo de uso:
numeros = [1, 2, 3, 4, 5]
resultado = calcular_estadisticas(numeros)
print(resultado)

numeros_vacios = []
resultado_vacio = calcular_estadisticas(numeros_vacios)
print(resultado_vacio)
```

9. Escribir una función que calcule el máximo común divisor de dos números y otra que calcule el mínimo común múltiplo.

```python
def mcd(n, m):
  while(m):
    # Intercambia los valores de n y m
    # n toma el valor de m
    # m toma el valor del resto de la división de n entre m
    n, m = m, n % m
  return n

def mcm(n, m):
  # El MCM se calcula multiplicando los dos números y dividiendo el resultado por su MCD
  # Usamos int() para asegurar que el resultado sea un entero
  return abs(n*m) // mcd(n, m)

# Ejemplo de uso
numero1 = 48
numero2 = 18

print(f"El MCD de {numero1} y {numero2} es: {mcd(numero1, numero2)}")
print(f"El MCM de {numero1} y {numero2} es: {mcm(numero1, numero2)}")
```

10. Escribir una función que convierta un número decimal en binario y otra que convierta un número binario en decimal.

```python
def decimal_a_binario(n):
    if n == 0:
        return "0"
    binario = ""
    while n > 0:
        binario = str(n % 2) + binario
        n //= 2  # División entera
    return binario
```

11. Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.

```python
def count_words(text):
     text = text.split()
    words = {}
    for i in text:
        if i in words:
            words[i] += 1
        else:
            words[i] = 1
    return words

def most_repeated(words):
    max_word = ''
    max_freq = 0
    for word, freq in words.items():
        if freq > max_freq:
            max_word = word
            max_freq = freq
    return max_word, max_freq

text = 'Como quieres que te quiera si el que quiero que me quiera no me quiere como quiero que me quiera'
print(count_words(text))
print(most_repeated(count_words(text)))
```
